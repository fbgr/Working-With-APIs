# Reading nobel prizes

```python
response = requests.get('https://api.nobelprize.org/2.1/nobelPrizes')
response
```

```python
results = response.json()['nobelPrizes']
```

```python
from pandas import json_normalize
results = response.json()['nobelPrizes']
flattened_data = json_normalize(results)
flattened_data.head(5)

# It doesn't work for links and laureates (I think because we have '[]')
# I'll try to fix it because i want to see their names
```

```python
# removing [] from the columns
for i,item in enumerate(flattened_data['links']):
    flattened_data['links'][i] = item[0]

for i,item in enumerate(flattened_data['laureates']):
    flattened_data['laureates'][i] = item[0]
flattened_data.head(5)
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
# flattening with the function we saw in class: links and laureates
nobel_data = flatten(flattened_data,['links','laureates'])
```
